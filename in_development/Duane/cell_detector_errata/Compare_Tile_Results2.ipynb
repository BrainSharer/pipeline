```python
#MANUAL TEST OF CELL CANDIDATE EXTRACTION (PHYSICAL TILES): (with filter)

import cv2
import numpy as np
from pathlib import Path
import imageio.v2 as imageio
import matplotlib.pyplot as plt

#CONSTANTS:
segmentation_threshold = 2000


def show_image(img_array):
    plt.imshow(img_array)
    plt.show()

def load_image(file):
    return imageio.imread(file)


def subtract_blurred_image(image):
    """
    STEP INVOLVES COMPARING EACH TILE IN BOTH CHANNELS AND FINDING DIFFERENCES

    average the image by subtracting gaussian blurred mean
    """
    small = cv2.resize(image, (0, 0), fx=0.05, fy=0.05, interpolation=cv2.INTER_AREA)
    blurred = cv2.GaussianBlur(small, ksize=(21, 21), sigmaX=10)
    relarge = cv2.resize(blurred, image.T.shape, interpolation=cv2.INTER_AREA)
    difference = image - relarge
    return difference


def find_connected_segments(image, segmentation_threshold) -> tuple:
    """find connected segments (cell candidates)"""
    n_segments, segment_masks, segment_stats, segment_location = cv2.connectedComponentsWithStats(np.int8(image > segmentation_threshold))
    segment_location = np.int32(segment_location)
    segment_location = np.flip(segment_location, 1)

    return (n_segments, segment_masks, segment_stats, segment_location)


def get_cell_candidates(animal, section, connected_segments, max_segment_size, cell_radius, x_window, y_window, row, col, difference_ch3, difference_ch1):
        """creates examples for one tile

        :type tile: _type_
        :return: examples
        :rtype: _type_

        connected_segments: tuple - contains information about cell candidates, including dimensions of bounding boxes around cell candidates
        
        Area is for the object, where pixel values are not zero
        """
        
        n_segments, segment_masks, segment_stats, segment_location = connected_segments
        print(f'segment_stats: {len(segment_stats)}')
        #segment_stats, segment_location may not be required if we are passing abosolute coordinates

        #print(f'DEBUG - n_segments, segment_masks, segment_stats, segment_location: {n_segments, segment_masks, segment_stats, segment_location}')
        Examples = []
        for segmenti in range(n_segments):
            #_, _, candidate_bounding_box_width, candidate_bounding_box_height, candidate_bounding_box_area = segment_stats[segmenti, :]
            #print(f'candidate_bounding_box_width, candidate_bounding_box_height, candidate_bounding_box_area: {candidate_bounding_box_width, candidate_bounding_box_height, candidate_bounding_box_area}')
            _, _, width, height, area = segment_stats[segmenti, :]
            #print(f'DEBUG - area: {area}')

            if area > max_segment_size:
                continue
            segment_row, segment_col = segment_location[segmenti, :]
            row_start = int(segment_row - cell_radius)
            col_start = int(segment_col - cell_radius)
            if row_start < 0 or col_start < 0:
                continue
            row_end = int(segment_row + cell_radius)
            col_end = int(segment_col + cell_radius)
            
            if row_end > y_window or col_end > x_window: #do we really need to check for this?
                continue

            segment_mask = segment_masks[row_start:row_end, col_start:col_end] == segmenti
            example = {'animal': animal,
                       'section': section,
                       'area': area,
                       'row': segment_row * row, #need to be absolute coordinates
                       'col': segment_col * col, #need to be absolute coordinates
                       'height': height,
                       'width': width,
                       'image_CH3': difference_ch3[row_start:row_end, col_start:col_end],
                       'image_CH1': difference_ch1[row_start:row_end, col_start:col_end],
                       'mask': segment_mask}
            
            Examples.append(example)

        return Examples

animal = 'DK52'
section = '006'
cell_radius = 40
max_segment_size = 100000
input_path_virus = Path(f'/net/birdstore/Active_Atlas_Data/cell_segmentation/{animal}/CH3/{section}/')
input_path_dye = Path(f'/net/birdstore/Active_Atlas_Data/cell_segmentation/{animal}/CH1/{section}/')

# input_file_virus1 = Path(f'/net/birdstore/Active_Atlas_Data/data_root/pipeline_data/DK52/preps/C3/full_aligned/', str(section) + '.tif')
# bigimg = load_image(input_file_virus1)
# print(f'VIRUS bigimg (CH1): SHAPE: {bigimg.shape}')
# input_file_dye1 = Path(f'/net/birdstore/Active_Atlas_Data/data_root/pipeline_data/DK52/preps/C1/full_aligned/', str(section) + '.tif')
# bigimg = load_image(input_file_dye1)
# print(f'DYE bigimg (CH3): SHAPE: {bigimg.shape}')

total_cell_candidates=0
connected_segments_total=0
for tile in range(10):
    input_file = Path(input_path_virus, f'{section}tile-{tile}.tif')
    image_roi_virus = load_image(input_file)
    input_file = Path(input_path_dye, f'{section}tile-{tile}.tif')
    image_roi_dye = load_image(input_file)
    
    #show_image(image_roi_virus)

    print(f'IMAGE: {section}, tile:{tile}, channel: 3/virus, SHAPE: {image_roi_virus.shape}')
    print(f'IMAGE: {section}, tile:{tile}, channel: 1/dye, SHAPE: {image_roi_dye.shape}')
    

    # #CALCULATE DIFFERENCE BETWEEN VIRUS (e.g. FLUORESCENCE) AND DYE (e.g. NEUROTRACE)
    difference_ch3 = subtract_blurred_image(image_roi_virus)
    difference_ch1 = subtract_blurred_image(image_roi_dye)

    # #FIND CONNECTED SEGMENTS (CELL CANDIDATES)
    connected_segments = find_connected_segments(difference_ch3, segmentation_threshold)
    print(f'find_connected_segments: {connected_segments[0]}')
    connected_segments_total += connected_segments[0]

    if connected_segments[0] > 2: #FOUND CELL CANDIDATE
        #print('FOUND CELL CANDIDATE') #TODO: REMOVE
        cell_candidates = get_cell_candidates(animal, section, connected_segments, max_segment_size, cell_radius, x_window, y_window, row, col, difference_ch3, difference_ch1)
        total_cell_candidates += len(cell_candidates)
        print(f'cell_candidates: {len(cell_candidates)}')

print(f'raw connected_segments_total: {connected_segments_total}')
print(f'total_cell_candidates: {total_cell_candidates}')   
```
